```python
df = pd.read_csv(r"/kaggle/input/bank-marketing-data-set/bank/bank-full.csv",sep=";")
df.head()
```

```python
cols = ["age","job","marital","education","balance","housing","contact","day","month","duration","campaign","pdays","previous","poutcome","y"]
df = df[cols]
df.head()
```

```python
numerical_features = [col for col in df.columns if df[col].dtype=="int64"]
numerical_features
```

```python
df[numerical_features].corr()
```

```python
df['y'] = df['y'].replace({'yes': 1, 'no': 0})
df.head()
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.20, random_state=42)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.y.values
y_test = df_test.y.values
y_val = df_val.y.values

del df_train["y"]
del df_test["y"]
del df_val["y"]
```

```python
from sklearn.metrics import mutual_info_score
def mutual_info_y_score(series):
    return mutual_info_score(series, df_full_train.y)

categorical = [col for col in df.columns if df[col].dtype=="O"]
mi = df_full_train[categorical].apply(mutual_info_y_score)
mi.sort_values(ascending=False)
```

```python
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=False)
train_dict = df_train[categorical + numerical_features].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical + numerical_features].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)

```

```python
model.fit(X_train, y_train)
y_pred = model.predict_proba(X_val)[:, 1]
y_decision = (y_pred >= 0.5)
(y_val == y_decision).mean()
```

```python
for c in [0.01, 0.1, 1, 10, 100]:
    model = LogisticRegression(solver='liblinear', C=c, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict_proba(X_val)[:, 1]
    y_decision = (y_pred >= 0.5)
    print(c, "--", round((y_val == y_decision).mean(),3))

        
```
